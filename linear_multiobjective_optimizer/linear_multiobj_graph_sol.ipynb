```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams["figure.figsize"] = (10,10)
```

```python
plt.xlim((-1,10))
plt.ylim((-1,10))
plt.plot([-10, 10], [0, 0], color='black')
plt.plot([0, 0], [-10, 10], color='black')
x = np.linspace(-10, 10)
y1 = (8-4*x)/8
plt.plot(x, y1, color='C2',label="y1")
y2 = (6-3*x)/-6
plt.plot(x, y2, color='C3',label="y2")
y3 = (14-4*x)/-2
plt.plot(x, y3, color='C4',label="y3")
y4 = 1000*x-6000
plt.plot(x,y4,color='C5',label="y4")
y5 = (15+x)/3
plt.plot(x, y5, color='C6',label="y5")
y6 = (18+2*x)/4
plt.plot(x, y6, color='C7',label="y6")
y7 = (9+6*x)/3
plt.plot(x, y7, color='C8',label="y7")
plt.legend(loc="upper right")

A1 = np.array([[3, -6], [4, -2]])
B1 = np.array([6, 14])
X1 = np.linalg.solve(A1,B1)
A2 = np.array([[4, -2], [1, 0]])
B2 = np.array([14, 6])
X2 = np.linalg.solve(A2,B2)
A3 = np.array([[1, 0], [-1, 3]])
B3 = np.array([6, 15])
X3 = np.linalg.solve(A3,B3)
A4 = np.array([[-1, 3], [-2, 4]])
B4 = np.array([15, 18])
X4 = np.linalg.solve(A4,B4)
A5 = np.array([[-2, 4], [-6, 3]])
B5 = np.array([18, 9])
X5 = np.linalg.solve(A5,B5)
X1 = [2,4,6,6.3,1,0,0]
X2 = [0,1,5,7,6,5,3,1]
data=np.array([[2,0],[4,1],[6,5],[6,7],[3,6],[1,5],[0,3],[0,1]])
plt.fill(*data.T,label = 'Feasible region')
plt.scatter(*data.T)
for xy in zip(*data.T):                                       
    plt.annotate('(%s, %s)' % xy, xy=xy, textcoords='data') 
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Feasible region in decision space')
plt.legend()
plt.show()
```

```python
theta_z1 = np.array([[3],[-2]])
z1 = np.dot(data,theta_z1)
z1_loop = np.append(z1,z1[0:1],axis=0)
theta_z2 = np.array([[-1],[2]])
z2 = np.dot(data,theta_z2)
z2_loop = np.append(z2,z2[0:1],axis=0)
table = pd.DataFrame(data) 
table['Max Z1'] = z1
table['Max Z2'] = z2
table.index = ['A','B' , 'C','D','E','F','G','H']
table.columns = ['x1', 'x2', 'Max Z1','Max Z2']
print(table)
plt.xlim((-10,12))
plt.ylim((-10,10))
plt.plot([-10, 10], [0, 0], color='black')
plt.plot([0, 0], [-10, 10], color='black')
plt.fill(z1_loop,z2_loop)
plt.scatter(z1,z2)
for xy in zip(z1, z2):                                       
    plt.annotate('(%s, %s)' % xy, xy=xy, textcoords='data') 
plt.xlabel('Z1')
plt.ylabel('Z2')
plt.title('Feasible Solutions for objectives Z1 and Z2')
```
